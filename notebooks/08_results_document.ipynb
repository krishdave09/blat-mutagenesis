# 08 - benchmark results document

Assembles `docs/05_benchmark_results.md` and renders `docs/05_benchmark_results.pdf`.

The markdown is the canonical, curated document, committed to the repo. This notebook just re-renders the PDF from that markdown, embedding the five figures. PDF rendering needs `markdown` and `xhtml2pdf`, which live in the `python` env.

```python
import sys
from pathlib import Path
ROOT = next(p for p in [Path.cwd(), *Path.cwd().parents] if (p/'.projectroot').exists())
sys.path.insert(0, str(ROOT))
from paths import TABLES, FIGURES, DOCS
print("project root:", ROOT)
```

```python
import subprocess, sys, os
def run_in(env, script, *args):
    """Run a src/ script inside a named conda env, streaming output."""
    py = Path(sys.executable).parent.parent.parent / env / "bin" / "python"
    if not py.exists():  # fallback: same interpreter
        py = sys.executable
    cmd = [str(py), str(ROOT/"src"/script), *args]
    print("running:", " ".join(cmd))
    r = subprocess.run(cmd, cwd=str(ROOT/"notebooks"), capture_output=True, text=True)
    print(r.stdout[-3000:]); 
    if r.returncode!=0: print("STDERR:", r.stderr[-2000:]); raise SystemExit(r.returncode)
```

## Render the PDF from the canonical markdown
`make_benchmark_doc.py` resolves the figure references and produces the PDF. Run it in the `python` env, which has the PDF toolchain.

```python
run_in('python','make_benchmark_doc.py')  # -> docs/05_benchmark_results.pdf
```
